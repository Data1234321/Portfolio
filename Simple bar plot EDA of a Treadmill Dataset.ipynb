# Simlple bar plot EDA of a Treadmill Dataset

```python
import pandas as pd                   
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patheffects import withStroke
import plotly.express as px

import warnings
warnings.filterwarnings("ignore")
```

```python
cardio_good_fitness = pd.read_csv("/kaggle/input/treadmill-users-dataset/treadmil-users.csv")
```

## General user EDA

```python
product_sales = cardio_good_fitness[['Product']].value_counts().reset_index()
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple']

product_sales.plot(kind='bar', 
                   x='Product', 
                   y='count', 
                   xlabel='Product', 
                   ylabel='Sales', 
                   rot=0.5,
                   legend=False,
                   stacked=True,
                   color=colors  # Assign different colors to each bar
                  )

plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data=cardio_good_fitness, x=cardio_good_fitness['Fitness'], bins=5, legend=True, hue='Fitness')
plt.title('Distribution of fitness (All Products)')
plt.show();
```

```python
plt.figure(figsize=(6, 6))
sns.histplot(data=cardio_good_fitness, x='Gender', binwidth=1, hue='Gender', palette='dark')
plt.title('Genders')
plt.show();
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data=cardio_good_fitness, x=cardio_good_fitness['MaritalStatus'], binwidth=1, hue='MaritalStatus')
plt.title('Relationship Status')
plt.show();
```

## Ages of people that use certain treadmills

```python
TM798_product = cardio_good_fitness[cardio_good_fitness['Product'] == 'TM798']
TM195_product = cardio_good_fitness[cardio_good_fitness['Product'] == 'TM195']
TM498_product = cardio_good_fitness[cardio_good_fitness['Product'] == 'TM498']
```

```python
def treadmill_ages(data, column, color):
    plt.figure(figsize=(6,6))
    visualisation = sns.histplot(data=data, x=data['Age'], color=color, kde=True)
    plt.show(visualisation);
```

```python
treadmill_ages(cardio_good_fitness, 'Age', 'red')
```

```python
treadmill_ages(TM195_product, 'Age', 'green')
```

```python
treadmill_ages(TM498_product, 'Age', 'purple')
```

```python
treadmill_ages(TM798_product, 'Age', 'blue')
```
